```python
#Loading the required libraries
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
#loading the ipl matches dataset
ipl= pd.read_csv('IPL_matches.csv')
```

```python
#having a glance at the first five records of the dataset
ipl.head()
```

```python
#looking at the number of rows and columns in the datset
ipl.shape
```

```python
#Getting the frequency of the most man of the match awards
ipl['Player_of_Match'].value_counts()
```

```python
#Getting the top 10 players with most man of the match awards
ipl['Player_of_Match'].value_counts()[0:5]
```

```python
list(ipl['Player_of_Match'].value_counts()[0:5].keys())
```

```python
plt.figure(figsize=(8,5))
plt.bar(list(ipl['Player_of_Match'].value_counts()[0:5].keys()),list(ipl['Player_of_Match'].value_counts()[0:5]),color='g')
plt.show()
```

```python
#Getting the frequency of result column
ipl['WinningTeam'].value_counts()
```

```python
#Finding out numbers of toss win w.r.t each team
ipl['TossWinner'].value_counts()
```

```python
#Extracting the records where a team won batting-first
batting_first=ipl[ipl['WonBy'] == 'Runs' ]
```

```python
#Looking at the batting-first
batting_first.head()
```

```python
#Making a histogram
plt.figure(figsize=(5,7))
plt.hist(batting_first['Margin'])
plt.title('Distribution of Runs')
plt.xlabel('Runs')
plt.show()
```

```python
#Include out the number of wins w.r.t each time after batting-first
batting_first['WinningTeam'].value_counts()
```

```python
#making a bar_plot for top 3 teams with most wins after batting first
plt.figure(figsize=(8,6))
plt.bar(list(batting_first['WinningTeam'].value_counts()[0:3].keys()),list(batting_first['WinningTeam'].value_counts()[0:3]),color=["blue","yellow","red"])
```

```python
#Making a pie chart
plt.figure(figsize=(7,7))
plt.pie(list(batting_first['WinningTeam'].value_counts()),labels=list(batting_first['WinningTeam'].value_counts().keys()),autopct='%0.1f%%')
```

```python
#Extracting those records where a team has won after batting-second
batting_second=ipl[ipl['WonBy']!='Runs']
```

```python
#Looking at the head
batting_second.head()
```

```python
#Making histogram for frequency of wins w.r.t number of wickets
plt.figure(figsize=(5,5))
plt.hist(batting_second['Margin'],bins=30)
plt.show()
```

```python
#Finding out the frequency of numbers of wins w.r.t each time after batting-second
batting_second['WinningTeam'].value_counts()
```

```python
#Making a bar plot for top 3 teams with most wins after batting-second
plt.figure(figsize=(7,7))
plt.bar(list(batting_second['WinningTeam'].value_counts()[0:3].keys()),list(batting_second['WinningTeam'].value_counts()[0:3]),color=["purple","blue","red"])
plt.show()
```

```python
#Creating pie chart for distribution of most wins after batting second
plt.figure(figsize=(7,7))
plt.pie(list(batting_second['WinningTeam'].value_counts()),labels=list(batting_second['WinningTeam'].value_counts().keys()),autopct='%0.1f%%')
plt.show()
```
